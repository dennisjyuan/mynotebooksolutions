## How could you have improved performance?

*Hint:* Try plotting the data in three dimensions along with the hyperplane solution, and see if you can infer
a new variable which will help, or try a nonlinear/non-parametric model

````python
X_train_new = pd.DataFrame ({'TV':X_train['TV'], 'radio':X_train['radio']})
X_test_new = pd.DataFrame ({'TV':X_test['TV'], 'radio':X_test['radio']})


from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt



regr1 = LinearRegression()
regr1.fit(X_train_new, y_train)

print ('Coefficients: \n', regr1.coef_)
print ('Residual sum of squares = {:.5f}'.format(np.mean(regr1.predict(X_test_new) - y_test)**2))
print ('Variance score = {:.5f}'.format( regr1.score(X_test_new, y_test)))
print ('Variance score with newspaper data = {:.5f}'. format(regr.score(X_test, y_test)))


#plotting scatter with axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter (df['TV'], df['radio'], y, zdir ='z')
ax.set_xlabel('TV')
ax.set_ylabel('Radio')
ax.set_zlabel('Sales')


#setting up hyperplane 
c_tv = regr1.coef_[0]
c_radio = regr1.coef_[1]

normal = [c_tv, c_radio, -1]
d = regr1.intercept_

                     
#set up z coordinates
xx, yy = np.meshgrid(range(300), range (50))
z = (-normal[0]*xx - normal[1]*yy - d)*1./normal[2]



#set up 3D plot
plt3d = plt.figure().gca(projection ='3d')

#plot surface
plt3d.plot_surface(xx,yy,z,alpha=0.2)

#plot scatter
plt3d.scatter (df['TV'], df['radio'], y, zdir ='z')

plt3d.set_xlabel('TV')
plt3d.set_ylabel('Radio')
plt3d.set_zlabel('Sales')


````

````python
print ('Score without newspaper as a feature is higher than score with, indicating that having newspaper as a feature decreased model accuracy. By removing features that do not contribute, we reduce overfitting and improve model')
print ('By removing newspaper as a feature, we can construct a 3D plot with TV, radio, and sales as the dimensions and fit a hyperplane to the data, showing a good fit.')
````

````python
#setting data to higher p levels

pvalue = 2
X_train_p = pd.DataFrame ({'TV^2':X_train['TV']**pvalue, 'radio^2':X_train['radio']**pvalue})
X_test_p = pd.DataFrame ({'TV^2':X_test['TV']**pvalue, 'radio^2':X_test['radio']**pvalue})

X_test_p.head()

````

````python
X_train_p ['TV'] = X_train['TV']
X_train_p ['radio'] = X_train['radio']

X_test_p ['TV'] = X_test['TV']
X_test_p ['radio'] = X_test['radio']

X_train_p.head()
````

````python
#improving performance
#using p= pvalue


regr2 = LinearRegression()
regr2.fit(X_train_p, y_train)

print ('To try to improve performance, we can use p = 2 to improve fit and increase R^2')
print ('Coefficients: \n', regr2.coef_)
print ('Residual sum of squares = {:.5f}'.format(np.mean(regr2.predict(X_test_p) - y_test)**2))
print ('Variance score (p={}) = {:.5f}'.format (pvalue,regr2.score(X_test_p, y_test)))

print ('Variance score (p=1) = {:.5f}'.format (regr1.score(X_test_new, y_test)))

#improving performance
print ('\n')
print ('From the 3D fitting, I was not able to determine a new variable, but decided to expand the model by going from p=1 to p=2, generating 2 sets of features from the squared values of TV and radio.')
print ('The coefficients indicate that the impact of the squared inputs were relatively low compared to just TV and radio, but managed to imrpove the variance score.')

````

````python
# Regreesion tree

from sklearn import tree
clf = tree.DecisionTreeRegressor()
clf.fit(X_train_new, y_train)
clf.predict(X_test_new)
print('Reression tree score (p=1) = %.4f' % clf.score(X_test_new, y_test))

clfp = tree.DecisionTreeRegressor()
clfp.fit(X_train_p, y_train)
clfp.predict(X_test_p)
print('Reression tree score (p=2) = %.4f' % clfp.score(X_test_p, y_test))

print('\n')
print('Answer: Utilizing a regression tree, we see that p=1 gives a higher score than p=2.')
print('This coudl be due to extra features in p=2 that generates overfitting/oversplitting.')
print('Regression tree analysis provides higher scores and a better model for predictions for this data set.')
````

## Problem 3:  Gradient Descent and the learning rate

By modifying the learning rate below, show how the convergence takes longer or doesn't converge at all.
Can you explain in words or math why this is?

````python
from numpy import *

# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

def run(num_iterations):
    points = genfromtxt("data/data.csv", delimiter=",")
    learning_rate = 0.0001
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    print ("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print ("Running...")
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print ("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1])
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')



        
run(100)
````

````python
def run_learn(num_iterations, step, step_rate):
    points = genfromtxt("data/data.csv", delimiter=",")
    step_rate = step_rate
    learning_rate = step_rate*(step)
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    print ("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print ("Running... with learning rate = {}".format(step_rate*step))
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print ("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1])
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')


#changing learning rate over range 7 orders of magnitude
step_r = 0.00005
for rate in arange (0, 10):        
    run_learn(100, rate, step_r)
    plt.show()
````

````python
#discussion
print ('Answer: At very low learning rates, there is no convergence unless a very high number of iterations is used.')
print ('As learning rate increases, convergence rate increases until a critical value of learning rate, in which the solution does not converge.')
print ('This critical value for learning rate is when the model completely misses the exact point.')
````

````python
import numpy as np
def run_get_error(num_iterations, step, step_rate):
    points = genfromtxt("data/data.csv", delimiter=",")
    step_rate = step_rate
    learning_rate = step_rate*(step)
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations

    #print ("Running... with learning rate = {}".format(10**l_rate))
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    #print ("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    error = compute_error_for_line_given_points(b, m, points)
    return error

rate0 = 0
rate1 = 1
rate2 = 2
rate3 = 5
rate4 = 8
iteration = 20
step_r1 = 0.00005
e0 = []
e1 = []
e2 = []
e3 = []
e4 = []


for num in range(0,iteration):
    e0.append(run_get_error(num, rate0, step_r1))

for num in range(0,iteration):
    e1.append(run_get_error(num, rate1, step_r1))

for num in range(0,iteration):
    e2.append(run_get_error(num, rate2, step_r1))


for num in range(0,iteration):
    e3.append(run_get_error(num, rate3, step_r1))

for num in range(0,iteration):
    e4.append(run_get_error(num, rate4, step_r1))
    

error = pd.DataFrame ({'rate0':e0,'rate1':e1, 'rate2':e2, 'rate3':e3, 'rate4':e4})    


error




````
